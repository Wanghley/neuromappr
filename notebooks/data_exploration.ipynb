## Imagined Movements

```python
df_imag_class_1 = pd.read_csv('../data/feaSubEImg_1.csv', header=None)
df_imag_class_2 = pd.read_csv('../data/feaSubEImg_2.csv', header=None)
```

```python
df_imag_class_1.head()
```

Each column represent a different trial and each row represent a different electrode position.

```python
df_imag_class_1.describe()
```

```python
df_imag_class_2.describe()
```

```python
import matplotlib.pyplot as plt

fig, axes = plt.subplots(1, 2, figsize=(10, 5))  # Create a 1x2 grid of subplots

# Plot on the first subplot
axes[0].plot(df_imag_class_1.iloc[:, 0].T)
axes[0].set_title("Channel 1 - Imagined Movement Class 1")
axes[0].set_xlabel("Trial")
axes[0].set_ylabel("Amplitude")

# Plot on the second subplot
axes[1].plot(df_imag_class_2.iloc[:, 0].T)
axes[1].set_title("Channel 1 - Imagined Movement Class 2")
axes[1].set_xlabel("Trial")
axes[1].set_ylabel("Amplitude")

plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```

```python
mean_class_1 = df_imag_class_1.mean()
mean_class_2 = df_imag_class_2.mean()

plt.plot(mean_class_1, label='Class 1')
plt.plot(mean_class_2, label='Class 2')
plt.legend()
plt.title("Mean Signal Across Electrodes")
plt.show()
```

```python
import seaborn as sns

corr = df_imag_class_1.corr()
sns.heatmap(corr, cmap='coolwarm', center=0)
plt.title("Channel Correlation (Class 1)")
plt.show()
```

```python
plt.figure(figsize=(14, 6))

plt.plot(df_imag_class_1.iloc[0, :])
plt.plot(df_imag_class_1.iloc[104, :])

plt.title("Activation of First 6 Electrode-Channel Pairs (Imagined Movement - Class 1)")
plt.xlabel("Trial")
plt.ylabel("Feature Activation")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
